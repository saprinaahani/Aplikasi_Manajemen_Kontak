```python
# Inisialisasi daftar kontak awal
daftar_kontak = []

# Fungsi untuk menambahkan kontak baru
def tambah_kontak(nama, nomor):
    kontak = {'nama': nama, 'nomor': nomor}
    daftar_kontak.append(kontak)
    print(f"Kontak '{nama}' dengan nomor '{nomor}' telah ditambahkan!")

# Fungsi untuk mencari kontak berdasarkan nama
def cari_kontak(nama):
    for kontak in daftar_kontak:
        if kontak['nama'] == nama:
            return kontak
    return None

# Fungsi untuk menampilkan semua kontak
def tampilkan_kontak():
    if len(daftar_kontak) == 0:
        print("Daftar kontak kosong.")
    else:
        print("Daftar Kontak:")
        for kontak in daftar_kontak:
            print(f"Nama: {kontak['nama']}, Nomor: {kontak['nomor']}")

# Loop utama program
while True:
    print("\nAplikasi Manajemen Kontak")
    print("1. Tambah Kontak")
    print("2. Cari Kontak")
    print("3. Tampilkan Semua Kontak")
    print("4. Keluar")

    pilihan = input("Pilih operasi (1/2/3/4): ")

    if pilihan == '1':
        nama = input("Masukkan nama kontak: ")
        nomor = input("Masukkan nomor kontak: ")
        tambah_kontak(nama, nomor)
    elif pilihan == '2':
        nama = input("Masukkan nama kontak yang ingin dicari: ")
        kontak = cari_kontak(nama)
        if kontak:
            print(f"Nama: {kontak['nama']}, Nomor: {kontak['nomor']}")
        else:
            print(f"Kontak dengan nama '{nama}' tidak ditemukan.")
    elif pilihan == '3':
        tampilkan_kontak()
    elif pilihan == '4':
        print("Terima kasih telah menggunakan aplikasi Manajemen Kontak. Sampai jumpa!")
        break
    else:
        print("Pilihan tidak valid. Silakan pilih operasi yang sesuai (1/2/3/4).")
```
